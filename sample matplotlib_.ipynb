```python
import matplotlib.pyplot as plt
```

```python
lables=('Java', 'Python', 'PHP', 'C#', 'Scala')
index=(1, 2, 3, 4, 5)
sizes=[50, 39, 45, 30, 22]
```

```python
plt.bar(index, sizes, tick_label=lables,color=('green','yellow','red','blue','black'))
plt.ylabel('Usage')
plt.xlabel('Programing Languages')
plt.show()
```

```python
plt.pie(sizes,labels=lables,autopct='%1.f%%',counterclock=False,startangle=145)
plt.show()
```
